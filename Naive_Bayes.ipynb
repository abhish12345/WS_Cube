```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('placements.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python
sns.kdeplot(dataset["resume_score"])
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100, gnb.score(x_train,y_train)*100
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_train,y_train)*100, mnb.score(x_test,y_test)*100
```

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_train,y_train)*100, bnb.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
