```python
import pandas as pd
from sklearn import datasets
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=datasets.load_iris()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset.columns=['sepal_length','sepal_width','petal_length','petal_width']
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker='o')
plt.xlabel('no of cluster')
plt.xticks([i for i in range(2,21)])
plt.ylabel('WCSS')
plt.grid(axis='x')
plt.show()
```

```python
kmn = KMeans(n_clusters = 3)
dataset["Predict"] = kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(dataset,labels=kmn.labels_)
```

```python
ss = []
no_c = [j for j in range(2,21)]
for i in range(2,21):
    km1 = KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c,ss)
plt.xlabel('no of clusters')
plt.ylabel('silhouette_score')
plt.xticks(no_c)
plt.grid(axis='x')
plt.show()
```
