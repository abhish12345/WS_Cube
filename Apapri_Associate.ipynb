```python
market= []
for i in range(0,dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if type(dataset[j][i])==str:
            cus.append(dataset[j][i])
    market.append(cus)
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
p=collections.Counter(l)
```
